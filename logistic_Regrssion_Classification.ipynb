```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import Dataset

ds = pd.read_csv("Social_Network_Ads.csv")
ds
```

```python
ds.columns
```

```python
ds.info()
```

```python
#Find number of duplicates

ds.duplicated().sum()
```

```python
#Remove Dulicates

ds.drop_duplicates(inplace = True)
```

```python
ds.isnull().sum()  # Used to check null values are present or not
```

```python
#Splitting dataset into independent and dependent data

x =ds[['Age','EstimatedSalary']]
y=ds[['Purchased']]
```

```python
x
```

```python
y
```

```python
# Feature scaling   -To convert larger data value into smaller part

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x=pd.DataFrame(sc.fit_transform(x),columns=['Age','EstimatedSalary'])
```

```python
# Splitting the dataset into training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.1,random_state =1)
```

```python
x_test
```

```python
# applying the logistic regrssion model on the Training set

from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
#Prediction for myinput

myinput = pd.DataFrame(sc.transform([[30,150000]]),columns = ['Age','EstimatedSalary'])
classifier.predict(myinput)
```

```python
# Checking Accuracy of Logistic Regrssion model

from sklearn.metrics import accuracy_score

#Predict on the test set
y_pred = classifier.predict(x_test)

#Calculate Accuracy
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy of the logistic regression model:{accuracy*100:.2f}")
```

```python
# User Input

age = float(input("Enter age:"))
salary = float(input("Enter EstimatedSalary:"))

myinput = pd.DataFrame(sc.transform([[age,salary]]),columns =[ ['Age','EstimatedSalary']])

result = classifier.predict(myinput)

if result[0] == 1:
    print("Prediction : user will purchase plan")
else :
    print("Prediction : user will not purchase plan")
```
